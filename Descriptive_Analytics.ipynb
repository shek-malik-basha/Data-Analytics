```python
import statistics
```

```python
#Data Set
member_names=['Tharun','Zubeda','Shavkath','Sneha','Yagnavi']
age=[21,19,25,23,36]
workout_hours=[2,1,5,3,2]
Calories_Burned=[500,400,600,450,100]
print("Data set load sucessfully")
```

```python
#printing the data
for i in range(len(member_names)):
    print("Name:",member_names[i])
    print("Age:",age[i])
    print("Workout Hours:",workout_hours[i])
    print("Calories Burned:",Calories_Burned[i])
    print("...................")
```

```python
#mean of workouts
mean_workout_hours=statistics.mean(workout_hours)
print("Mean of Workout Hours:",mean_workout_hours)
```

```python
#Maxium Calories_Burned
max_Calories_Burned=max(Calories_Burned)
print("Maximum Calories Burned:",max_Calories_Burned)
```

```python
#Minimum Calories_Burned
min_Calories_Burned=min(Calories_Burned)
print("Minimum Calories Burned:",min_Calories_Burned)
```

```python
#Find Most Active person
most_calories_burned=max(Calories_Burned)
most_active_person_index=Calories_Burned.index(most_calories_burned)
most_active_person=member_names[most_active_person_index]
print("Most Active Person:",most_active_person)
```

```python
#find less Sctive person
less_calories_burned=min(Calories_Burned)
print("Less Active Person:",less_active_person)
```

```python
#Standard Devation
std_Calories_Burned=statistics.stdev(Calories_Burned)
print("Standard Devation of Calories Burned:",std_Calories_Burned)
```

```python
#Variance
var_Calories_Burned=statistics.variance(Calories_Burned)
print("Variance of Calories Burned:",var_Calories_Burned)
```

```python
import matplotlib.pyplot as plt
plt.plot(member_names,Calories_Burned)
plt.xlabel('member_names')
plt.ylabel('Calories Burned')
plt.title('Calories Burned by Members')
plt.show()
```

```python
#pie chart
plt.pie(Calories_Burned,labels=member_names,autopct='%1.1f%%')
plt.title('Calories Burned by Members')
plt.show()
```

```python
plt.scatter(member_names,Calories_Burned)
plt.xlabel('member_names')
plt.ylabel('Calories Burned')
plt.title('Calories Burned by Members')
plt.show()
```
