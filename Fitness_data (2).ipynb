```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read csv file
df = pd.read_csv('/content/fitness_tracker.csv')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
df_scaled.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_scaled = pd.DataFrame(df_scaled, columns=df.columns)
df_scaled.head()
```

```python
#one hot encoding
df_encoded = pd.get_dummies(df)
df_encoded.head()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
#find max calories burned
df_scaled['Calories_Burned'].max()
```

```python
df_scaled['Calories_Burned'].min()
```

```python
#mean
df_scaled['Calories_Burned'].mean()
```

```python
#most active person
max_calories_idx = df_scaled['Calories_Burned'].idxmax()
user_id_most_active = df.loc[max_calories_idx, 'User_ID']
print(f"The User ID with the most calories burned is: {user_id_most_active}")
```

```python
#most lazy person
min_calories_idx = df_scaled['Calories_Burned'].idxmin()
user_id_most_lazy = df.loc[min_calories_idx, 'User_ID']
print(f"The User ID with the most calories burned is: {user_id_most_lazy}")
```

```python
#standard devation
df_scaled['Calories_Burned'].std()
```

```python
#variance
df_scaled['Calories_Burned'].var()
```

```python
plt.bar(df['User_ID'], df_scaled['Calories_Burned'])
plt.xlabel('User ID')
plt.ylabel('Scaled Calories Burned')
plt.title('Scaled Calories Burned per User ID')
plt.show()
```

```python
plt.scatter(df_scaled['Calories_Burned'], df_scaled['Steps'])
plt.xlabel('Calories Burned')
plt.ylabel('Steps')
plt.title('Calories Burned vs Steps')
plt.show()
```
