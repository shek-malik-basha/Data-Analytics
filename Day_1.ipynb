```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Marks=[50,45,43,50,46,33,38,]
Study_hours=[8,6,5,8,6,4,5]
```

```python
data = {'Marks': Marks, 'Study Hours': Study_hours}
marks_study_df = pd.DataFrame(data)
display(marks_study_df.head())
```

```python
print(marks_study_df.mean())
print(marks_study_df.median())
print(marks_study_df.mode().T)
print(marks_study_df.var())
print(marks_study_df.std())
print(marks_study_df['Marks'].corr(marks_study_df['Study Hours']))
```

```python
plt.plot(marks_study_df['Marks'],marks_study_df['Study Hours'])
plt.xlabel('Marks')
plt.ylabel('Study Hours')
plt.show()
```
